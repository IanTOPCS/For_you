```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
import os

# 設定中文字體
font = FontProperties(fname=os.environ['WINDIR']+'\\Fonts\\kaiu.ttf', size=16)

# data2 資料
data2 = np.array([['高', '低', '中', '低', '中'],
                  ['低', '中', '高', '高', '中'],
                  ['高', '中', '中', '低', '中']])

# 將 "高", "中", "低" 對應為數值 3, 2, 1
mapping = {'高': 3, '中': 2, '低': 1}
data2_numeric = np.vectorize(mapping.get)(data2)

# 年份與公司標籤
years = ['109', '110', '111', '112', '113']
companies = ['甲公司', '乙公司', '丙公司']

# 繪製摺線圖
plt.figure(figsize=(10, 6))
for i, company in enumerate(companies):
    plt.plot(years, data2_numeric[i], marker='o', label=company)

# 設定 Y 軸為 "高", "中", "低"
y_ticks = [1, 2, 3]
y_labels = ['低', '中', '高']
plt.yticks(y_ticks, y_labels, fontproperties=font)

# 標籤與圖例
plt.xlabel('年份', fontproperties=font)
plt.ylabel('業績表現', fontproperties=font)
plt.title('各公司業績表現趨勢', fontproperties=font)
plt.legend(loc='lower right', prop=font)
plt.grid(True)

# 顯示圖表
plt.tight_layout()
plt.show()
```
