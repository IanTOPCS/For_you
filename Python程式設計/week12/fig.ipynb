```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
import os

# 
font = FontProperties(fname=os.environ['WINDIR']+'\\Fonts\\kaiu.ttf', size=16)

# data2 
data2 = np.array([['', '', '', '', ''],
                  ['', '', '', '', ''],
                  ['', '', '', '', '']])

#  "", "", ""  3, 2, 1
mapping = {'': 3, '': 2, '': 1}
data2_numeric = np.vectorize(mapping.get)(data2)

# 
years = ['109', '110', '111', '112', '113']
companies = ['', '', '']

# 
plt.figure(figsize=(10, 6))
for i, company in enumerate(companies):
    plt.plot(years, data2_numeric[i], marker='o', label=company)

#  Y  "", "", ""
y_ticks = [1, 2, 3]
y_labels = ['', '', '']
plt.yticks(y_ticks, y_labels, fontproperties=font)

# 
plt.xlabel('', fontproperties=font)
plt.ylabel('', fontproperties=font)
plt.title('', fontproperties=font)
plt.legend(loc='lower right', prop=font)
plt.grid(True)

# 
plt.tight_layout()
plt.show()
```
