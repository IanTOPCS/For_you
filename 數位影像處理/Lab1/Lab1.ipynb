# 題目：列印99乘法表

# 步驟
1. 上官網載python
2. 安裝 jupyter
3. 打code
4. 交作業

```python
for i in range(1, 10):                              #從一數到九，為99乘法表的九個部分
    for j in range(1, 10):                          #從一數到九，為99乘法表一個部分內的九個結果
        print(f"{j} X {i} = {i*j}", end = '\t')     #印出數字相乘的結果，以tab方式對齊
    print()                                         #換行
```
