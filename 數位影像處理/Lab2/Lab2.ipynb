# 安裝套件：

```python
from PIL import Image               # 安裝 pillow
import numpy as np                  # 安裝 numpy
import cv2                          # 安裝 opencv
import matplotlib.pyplot as plt     # 安裝 matplotlib
```

# 產生反白影像：

```python
image = cv2.imread('./source/Pictrue.png')                                  # 載入圖片檔
cv2.imshow('Original Image', image)                                         # 輸出原本樣子

image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)                             # 將圖片轉變成灰階
image = 255-image                                                           # 將圖片顏色反轉

cv2.imshow('INV', image)                                                    # 輸出圖片

cv2.waitKey(0)                                                              # 等待隨意按鍵結束瀏覽
cv2.destroyAllWindows()                                                     # 將小視窗關掉
```

# 二值化：

```python
img = cv2.imread("./source/Pictrue.png")                                    # 開檔
cv2.imshow('Original Image', img)                                           # 輸出原始檔案
img= cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)                                  # 轉灰階
_, binaryImg = cv2.threshold(img, 120, 255,cv2.THRESH_BINARY)               # 將圖片二值化

cv2.imshow('Inverted Image', binaryImg)                                     # 輸出結果

cv2.waitKey(0)                                                              # 等待隨意鍵以接續程式
cv2.destroyAllWindows()                                                     # 關閉額外開啟的小視窗
```
